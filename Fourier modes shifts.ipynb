Prototype: NEID solar CCF: 47 modes corresponding to 94 CCF "pixels"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 3))
x = np.linspace(-np.pi, np.pi, 94*1000)
y = np.sin(x-np.pi/2)
y2 = np.sin(x*47-np.pi/2)
plt.xticks([])
plt.yticks([])
plt.plot(x, y, label='1st mode')
plt.plot(x, y2, label='47nd mode')
plt.legend()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 94)

for i in range(94):
    fig = plt.figure(figsize=(10, 1))
    n_mode = i
    def f(x):
        return np.sin(x*n_mode-np.pi/2)
    plt.plot(f(x)-f(x-0.001), 'o-')
    plt.title('mode = ' + str(n_mode) )
    plt.xticks([])
    plt.show()
```
