Prototype: NEID solar CCF: 47 modes corresponding to 94 CCF "pixels"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 3))
x = np.linspace(-np.pi, np.pi, 94*1000)
y = np.sin(x-np.pi/2)
y2 = np.sin(x*2-np.pi/2)
y47 = np.sin(x*47-np.pi/2)
plt.xticks([])
plt.yticks([])
plt.plot(x, y, label='1st mode')
plt.plot(x, y2, label='2st mode')
plt.plot(x, y47, label='47nd mode')
plt.legend()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 94)

for i in range(5):
    fig = plt.figure(figsize=(10, 1))
    n_mode = i
    def f(x):
        return np.sin(x*n_mode-np.pi/2)
    plt.plot(f(x), 'o--')
    plt.title('mode = ' + str(n_mode) )
    plt.xticks([])
    plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 94)

for i in range(48):
    fig = plt.figure(figsize=(10, 1))
    n_mode = i
    def f(x):
        return np.sin(x*n_mode-np.pi/2)
    plt.plot(f(x)-f(x-0.001), 'o--')
    plt.title('mode = ' + str(n_mode) + ' shifted')
    plt.xticks([])
    plt.show()
```

```python
fig = plt.figure(figsize=(15, 3))
n_mode = 34
def f(x, n_mode):
    return np.sin(x*n_mode-np.pi/2)
plt.plot(f(x, n_mode)-f(x-0.001, n_mode), 'o--', label=n_mode)
n_mode +=1
plt.plot(f(x, n_mode)-f(x-0.001, n_mode), 'o--', label=n_mode)
plt.title('mode = ' + str(n_mode) + ' shifted')
plt.xticks([])
plt.legend()
plt.show()

fig = plt.figure(figsize=(15, 3))
n_mode = 35
plt.plot( (f(x, n_mode)-f(x-0.001, n_mode)) - (f(x, n_mode+1)-f(x-0.001, n_mode+1)) , 'o--', label=n_mode)
plt.show()

fig = plt.figure(figsize=(15, 3))
n_mode = 35
plt.plot( (f(x, n_mode)-f(x-0.001, n_mode)) + (f(x, n_mode+1)-f(x-0.001, n_mode+1)) , 'o--', label=n_mode)
plt.show()
```
