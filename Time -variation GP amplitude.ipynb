```python
def GP(t, y, yerr):
    w0 = 2*np.pi/6
    Q = 10
    S0 = np.var(y) / (w0 * Q)
    bounds = dict(log_S0=(-15, 15), log_Q=(-15, 15), log_omega0=(-15, 15))
    kernel = terms.SHOTerm(log_S0=np.log(S0), log_Q=np.log(Q), log_omega0=np.log(w0), bounds=bounds)
    kernel.freeze_parameter("log_omega0")

    gp = celerite.GP(kernel, mean=np.mean(y))
    gp.compute(t, yerr)  # You always need to call compute once.

    initial_params = gp.get_parameter_vector()
    bounds = gp.get_parameter_bounds()

    r = minimize(neg_log_like, initial_params, method="L-BFGS-B", bounds=bounds, args=(y, gp))
    gp.set_parameter_vector(r.x)
    
    return gp
```

```python
S0 = 0.12
Q = 6
N = 3500

t = np.arange(N)/10
yerr = (np.random.normal(0.05, 0.03, N)**2 + 0.24**2)**0.5

alpha=0.6

plt.rcParams.update({'font.size': 14})

# Figure 2        
fig = plt.figure(figsize=(18, 3))
kernel = k1 = terms.SHOTerm(log_S0=np.log(S0), log_Q=np.log(Q), log_omega0=np.log(2*np.pi/(1/3033.754e-6/60)))
gp = celerite.GP(kernel)
gp.compute(t, yerr)
y = gp.sample()
plt.plot(t, y, alpha=alpha, label='total')

# _, _, _, _, t_sample, y_sample, yerr_sample = random_ts(t, y, yerr, test_ratio=0.1)
idx_array = (np.arange(int(N/10/1.75))*1.75*10)
idx = np.array([int(idx_i) for idx_i in idx_array])
t_sample, y_sample, yerr_sample = t[idx], y[idx], yerr[idx]
plt.errorbar(t_sample, y_sample, yerr_sample, fmt="k.", capsize=0, alpha=1, label='sample')

pred_mean, pred_var = GP(t_sample, y_sample, yerr_sample).predict(y_sample, t, return_var=True)
pred_std = np.sqrt(pred_var)        
plt.plot(t, pred_mean, color='r', alpha=0.3, label='gp prediction')
plt.fill_between(t, pred_mean+pred_std, pred_mean-pred_std, color='r', alpha=0.1,
                 edgecolor="none")
plt.xlabel("Time [min]")
plt.ylabel("RV [m/s]")
plt.legend()
plt.show()

# Figure 3
frequency, power = LombScargle(t_sample, y_sample).autopower(maximum_frequency=1, samples_per_peak=10)        
plot_x = 1/frequency
idx = (plot_x<8) 

# Periodogram
fig = plt.figure(figsize=(18, 3))
plt.plot(plot_x[idx], power[idx], 'k', alpha=0.5)
plt.ylim(bottom=0, top=max(power[idx])*1.05)    
plt.ylabel('Power')
plt.xlabel('Period [min]')
# plt.xscale('log')
plt.show()     


# Figure 4
fig = plt.figure(figsize=(18, 3))
plt.title('RV histogram')
_ = plt.hist(y_sample, bins=20, alpha=0.5)
plt.xlabel('RV [m/s]')
plt.show()


# Figure 5
fig = plt.figure(figsize=(18, 3))
plt.title('RV error histogram')
_ = plt.hist(yerr_sample, bins=20, alpha=0.5)
plt.xlabel('RV error [m/s]')
plt.show()    
    
```
